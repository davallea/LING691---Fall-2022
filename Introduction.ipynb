Population size = 100 000 000, mean = 60ms., and standard deviation = 8ms.
A data set with these characteristics is created and its mean is calculated. 
This represents the population parameter since every member of the population has been analyzed. 

```r
VOT_population <- rnorm(100000000, mean=60, sd=8)
mean(VOT_population)
hist(VOT_population)
abline(v=mean(VOT_population), col='red', lwd=3)
```

A random sample of 1000 individuals is drawn from the population and its mean is calculated. Notice that the mean sample statistic is very close to the population. However, the researcher does not know or have access to the population parameter in order to evaluate the results obtained from the population. 

```r
vot_sample <- sample(VOT_population, 1000, replace=TRUE)
mean(vot_sample)
hist(vot_sample)
abline(v=mean(vot_sample), col='red', lwd=3)
```

Now, a shorter random sample of 100 individuals is drawn. Now, we see a larger difference between the population parameter and the sample statistic. This difference is what is known as sampling error. 

```r
vot_sample <- sample(VOT_population, 100, replace=TRUE)
mean(vot_sample)
hist(vot_sample)
abline(v=mean(vot_sample), col='red', lwd=3)
```

If we were to drawn a 1000 random samples of 100 individuals each and plot their means against the population histogram. Most of these sample means approximate very well the population mean. 

```r
hist(VOT_population)
for (x in 1:1000) {
  vot_sample <- sample(VOT_population, 100, replace=TRUE)
  abline(v=mean(vot_sample), col='red', lwd=1)
}
```

A sample drawn less randomly in which not a random number generator was used. However, given the nature of R for creating the data set, this is still a reasonably random sample. 

```r
vot_sample <- VOT_population[5000000:5000100]
mean(vot_sample)
hist(vot_sample)
abline(v=mean(vot_sample), col='red', lwd=3)
```

However, a biased (not random) sample has a high likelihood of providing some distortion of the of the properties of the population. In this case, the sample statistic and the population parameter differ significantly.  

```r
vot_sample <- sort(VOT_population)[5000000:5000100]
mean(vot_sample)
hist(vot_sample)
abline(v=mean(vot_sample), col='red', lwd=3)
```
