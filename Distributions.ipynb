```r
# Sample data$ length
means <- c()
for (x in 1:10000) {
  sample_mean <- round(mean(sample(data$length, 1000, replace=TRUE)), digits=2)
    means <- append(means, sample_mean)
    }
```

```r
df<- data.frame(means)

graph <- ggplot(df, aes(x=means)) + 
    geom_bar() + 
    labs(title="Sample length frequencies", x="Word length", y="Frequency")

graph

#ggsave("word_length_sample_means.pdf", graph)
```
